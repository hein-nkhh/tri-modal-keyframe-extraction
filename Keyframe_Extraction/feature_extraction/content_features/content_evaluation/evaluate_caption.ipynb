```python
import pandas as pd
from clair_main import clair
import os
from tqdm import tqdm
tqdm.pandas()
```

```python
df=pd.read_csv('caption_phi.csv')
df
```

```python
df2=pd.read_csv('captions_comparison.csv')
df2
```

```python
df3=df2.drop(['prediction'],axis=1)
df3
```

```python
df4=df.merge(df3,on='image_id',how='inner')
df4
```

```python
dff1=df4[['phi_prediction','ref1']]
dff2=df4[['phi_prediction','ref2']]
dff3=df4[['phi_prediction','ref3']]
dff4=df4[['phi_prediction','ref4']]
dff5=df4[['phi_prediction','ref5']]
```

```python
# Include your OPENAI API KEY here
os.environ['OPENAI_API_KEY'] = ''
```

```python
def clair_scores(x,r):
    clair_score = clair.clair(x, r, model='gpt4')
    return clair_score[0]
```

```python
dff1['claire_score']=dff1.progress_apply(lambda row: clair_scores(row['phi_prediction'], row['ref1']), axis=1)
dff1
```

```python
dff2['claire_score']=dff2.progress_apply(lambda row: clair_scores(row['phi_prediction'], row['ref2']), axis=1)
dff2
```

```python
dff3['claire_score']=dff3.progress_apply(lambda row: clair_scores(row['phi_prediction'], row['ref3']), axis=1)
dff3
```

```python
dff4['claire_score']=dff4.progress_apply(lambda row: clair_scores(row['phi_prediction'], row['ref4']), axis=1)
dff4
```

```python
dff5['claire_score']=dff5.progress_apply(lambda row: clair_scores(row['phi_prediction'], row['ref5']), axis=1)
dff5
```

```python
dff1.to_csv('claire_evaluations/claire_output_1.csv')
dff2.to_csv('claire_evaluations/claire_output_2.csv')
dff3.to_csv('claire_evaluations/claire_output_3.csv')
dff4.to_csv('claire_evaluations/claire_output_4.csv')
dff5.to_csv('claire_evaluations/claire_output_5.csv')
```

```python
df_1=pd.read_csv('claire_evaluations/claire_output_1.csv')
df_2=pd.read_csv('claire_evaluations/claire_output_2.csv')
df_3=pd.read_csv('claire_evaluations/claire_output_3.csv')
df_4=pd.read_csv('claire_evaluations/claire_output_4.csv')
df_5=pd.read_csv('claire_evaluations/claire_output_5.csv')
```

```python
df=pd.concat([df_1['claire_score'],df_2['claire_score'],df_3['claire_score'],df_4['claire_score'],df_5['claire_score']],axis=1)
df
```

```python
df = df[~df.eq(0).any(axis=1)]
df
```

```python
df['score_mean']=df.mean(axis=1)
df
```
